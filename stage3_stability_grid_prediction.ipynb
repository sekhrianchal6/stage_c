```python
def fun(model, feat, col_name):
  #this function returns the importance of every feature
  i = pd.Series(model.feature_importances_, feat.columns).sort_values()
  i_df = pd.DataFrame(i).reset_index()
  i_df.columns = ['Features', col_name]
  i_df[col_name].round(3)
  return i_df
```
